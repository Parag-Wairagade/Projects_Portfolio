#  Investigating Hotel Business By Data Visualization

```python
# importing libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
# importing data

df = pd.read_csv("Hotel_Booking.csv")
```

```python
pd.set_option('display.max_columns',None)
```

```python
df.sample(10)
```

```python
df.info()
```

```python
# we need to change data type of children becuase it should be int not float

df.children = df.children.astype(int)
```

```python
df.info()
```

```python
# checking null values
df.isnull().sum()
```

```python
# we have to replace null values

df.children.fillna(0, inplace = True)
df.city.fillna('undefined', inplace = True)
df.agent.fillna(0, inplace = True)
df.company.fillna(0, inplace = True)
```

```python
df.isna().sum()
```

```python
# checking Duplicate values

df.duplicated().sum()
```

```python
# droping Duplicates

df.drop_duplicates(keep='first', inplace = True)
```

```python
# checking Duplicate values

df.duplicated().sum()
```

# Data Visulazation

```python
df_1 = df.groupby('hotel')[['hotel']].count()
df_1.columns = ['Total']
df_1.reset_index(inplace=True)
df_1
```

```python
plt.figure(figsize=(12,8))
sns.set()
plt.pie( df_1.Total, labels = df_1.hotel, autopct = '%.2f%%', radius = .8, startangle = 90, 
        wedgeprops = {'linewidth' : 3, 'edgecolor' : 'white'}, colors = ['g','b'])

plt.title('Total Hotel Bookings', fontdict = {'fontsize' : 20, 'fontweight' : 'bold'})
plt.legend(labelcolor = ['g','b'])
plt.show()
```

* **City Hotel** is booked more than **Resort Hotel** 

#  Ratio Of Monthly Booking

```python
df_2 = df.groupby(['hotel', 'arrival_date_month'])[['arrival_date_month']].count()
df_2.columns = ['Total']
df_2.reset_index(inplace=True)
df_2.sort_values('Total', ascending = False, inplace = True)
df_2
```

```python
plt.figure(figsize = (16,8))
sns.lineplot(x = df_2.arrival_date_month, y = df_2.Total, hue = df_2.hotel, style = df_2.hotel, palette = ['g','b'],
             dashes = False, markers = ["o","<"])

plt.title("Total Hotel Booking By Months", fontdict = {'fontsize' : 20, 'fontweight' : 'bold'})
plt.xlabel("Months", fontdict = {'fontsize' : 10, 'fontweight' : 'bold'})
plt.ylabel("Total Bookings", fontdict = {'fontsize' : 10, 'fontweight' : 'bold'})
plt.show()
```

* We can see that in **October** and **September** we got **Highest** number of Bookings.
* Also, **July** and **June** have **Moderate** number of Bookings.
* In **March** we got **Lowest** number of Bookings.

# Total Cancelation

```python
df_city = df[df.hotel == 'City Hotel']
df_resort = df[df.hotel == 'Resort Hotel']

c_city = df_city.groupby('is_canceled')[['is_canceled']].count()
c_city.columns = ['Total']
c_city.reset_index(inplace=True)

c_resort = df_resort.groupby('is_canceled')[['is_canceled']].count()
c_resort.columns = ['Total']
c_resort.reset_index(inplace=True)

x = {0 : 'Not Canceled', 1 : 'Canceled'}

c_city.is_canceled = c_city.is_canceled.map(x)
c_resort.is_canceled = c_resort.is_canceled.map(x)
```

```python
c_city
```

```python
c_resort
```

```python
plt.figure(figsize = (12,8))
plt.pie( c_city.Total, labels = c_city.is_canceled, autopct = '%.2f%%', radius = .8, startangle = 90, 
        wedgeprops = {'linewidth' : 3, 'edgecolor' : 'white'}, colors = ['g','b'])

plt.title('Total Hotel Bookings', fontdict = {'fontsize' : 20, 'fontweight' : 'bold'})
plt.legend(labelcolor = ['g','b'])
plt.show()
```

```python
plt.figure(figsize = (12,8))
plt.pie( c_resort.Total, labels = c_resort.is_canceled, autopct = '%.2f%%', radius = .8, startangle = 90, 
        wedgeprops = {'linewidth' : 3, 'edgecolor' : 'white'}, colors = ['g','b'])

plt.title('Total Hotel Bookings', fontdict = {'fontsize' : 20, 'fontweight' : 'bold'})
plt.legend(labelcolor = ['g','b'])
plt.show()
```

* In **City Hotels** canceletions rate is much **Higher** than **Resort Hotels**.

# Total Stay Duration


```python
df['total_stay_duration'] = df.stays_in_weekend_nights + df.stays_in_weekdays_nights
df
```

```python
df['total_stay_duration'].max()
```

```python
df['total_stay_duration'].min()
```

* So, the guests stayed betweem min. 0 to max 69 days.

```python
df['total_stay_week'] = np.where(df.total_stay_duration < 8, '<1',
                                 np.where(df.total_stay_duration < 15, '1-2',
                                 np.where(df.total_stay_duration < 22, '2-3',
                                 np.where(df.total_stay_duration < 29, '3-4', '>4'))))
df
```

```python
df_3 = df.groupby(['hotel','total_stay_week'])[['hotel']].count()
df_3.columns = ['Total']
df_3.reset_index(inplace = True)
df_3.sort_values('Total', ascending = False, inplace = True)
df_3
```

```python
plt.figure(figsize = (16,8))
sns.set()
sns.lineplot(x = df_3.total_stay_week, y = df_3.Total, hue = df_3.hotel, style = df_3.hotel, palette = ['g','b'],
             dashes = False, markers = ["o","<"])

plt.title("Total No Of Week Bookings", fontdict = {'fontsize' : 20, 'fontweight' : 'bold'})
plt.xlabel("Weeks", fontdict = {'fontsize' : 10, 'fontweight' : 'bold'})
plt.ylabel("Total Bookings", fontdict = {'fontsize' : 10, 'fontweight' : 'bold'})
plt.show()
```

* Most of the guests book hotels for not more than **1 Week**.
* Some of the guests book hotels for **1 to 2 Weeks**.
* Very few guests book hotels for  more than **2 Weeks**.
