# Data Visualization

```python
df_top5 = df.sort_values(['Total Number of Accidents 2013-2016'], ascending= False).head(5)
df_bottom5 = df.sort_values(['Total Number of Accidents 2013-2016'], ascending= True).head(5)
```

```python
plt.figure(figsize=(12,6))
plt.bar(df_top5['States/UTs'], df_top5['Total Number of Accidents 2013-2016'],width =0.4 , color = 'r')
plt.title('Top 5 Accidents State/Uts', fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.bar(df_bottom5['States/UTs'], df_bottom5['Total Number of Accidents 2013-2016'],width =0.4 , color = 'g')
plt.title('Bottom 5 Accidents State/Uts', fontsize=20)
plt.show()
```

* **Tamil Nadu** have the **Highest** number of accidents followed by **Maharashtra**.
* **Lakshadweep** have the **Lowest** number of accidents followed by **Daman & Diu**.

***Population plays major factor in number of accidents, we can see low populated states have low number of accidents.***

```python
df['Total Number of Road Accidents per Lakh Population - 2013-2016'] = np.mean([df['Total Number of Accidents Per Lakh Population - 2013'], df['Total Number of Accidents Per Lakh Population - 2014'], df['Total Number of Accidents Per Lakh Population - 2015'], df['Total Number of Accidents Per Lakh Population - 2016']], axis = 0)
```

```python
df
```

```python
df_top5 = df.sort_values(['Total Number of Road Accidents per Lakh Population - 2013-2016'], ascending= False).head(5)
df_bottom5 = df.sort_values(['Total Number of Road Accidents per Lakh Population - 2013-2016'], ascending= True).head(5)
```

```python
plt.figure(figsize=(12,6))
plt.bar(df_top5['States/UTs'], df_top5['Total Number of Road Accidents per Lakh Population - 2013-2016'],width =0.4 , color = 'r')
plt.title('Top 5 Accidents State/Uts by Population per Lakh', fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.bar(df_bottom5['States/UTs'], df_bottom5['Total Number of Road Accidents per Lakh Population - 2013-2016'],width =0.4 , color = 'g')
plt.title('Bottom 5 Accidents State/Uts by Population per Lakh', fontsize=20)
plt.show()
```

***Whene we considerd the Population  of the States we get differnt results***

* **Goa** have **High** accidents rate per population followed by **Kerla**.
* **Telangana** have the **Lowest** accidents rate per population followed by **Lakshadweep**.

```python
om = round(np.mean(df1['Offenders (Driver/Pedestrian) Died_Male']),2)
of = round(np.mean(df1['Offenders (Driver/Pedestrian) Died_Female']),2)
ot = np.mean(df1['Offenders (Driver/Pedestrian) Died_Transgender'])
om,of,ot
```
